```python
import pandas as pd
```

```python

fuel = pd.read_csv('fuel.csv', usecols=['Vehicle ID', 'Year', 'Make', 'Model', 'Class', 'Fuel Type', 'Combined MPG (FT1)'], low_memory=False, index_col='Vehicle ID')
fuel.head()
```

```python
fuel['Combined KPL'] = fuel['Combined MPG (FT1)'] * 1.609 / 3.78
```

```python
fuel
```

```python
fuel.insert(loc=5, column = 'liters per km', value = 100 / fuel["Combined KPL"])
```

```python
del fuel["Combined MPG (FT1)"]
```
