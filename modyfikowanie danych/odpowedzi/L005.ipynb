```python
import pandas as pd
import numpy as np

fortune = pd.read_csv('Fortune_500_2017.csv', usecols=['Rank', 'Title', 'Industry', 'Hqlocation', 'Sector'], index_col='Rank')
fortune.head()
```

```python
fortune["Sector"] = fortune["Sector"].str.upper()
fortune.head()
```

```python
fortune[fortune["Industry"].str.lower().str.contains('comp')].head()
```

```python
fortune[['City','State']] = fortune["Hqlocation"].str.split(",",expand=True)
fortune.head()
```

```python
def BuildShortcut(row):
    industry = row["Industry"]
    result = ''
    for i in industry.split(' '):
        result += i[0]
    return result
```

```python
BuildShortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(BuildShortcut,axis=1)
fortune.head()
```
