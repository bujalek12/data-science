```python
import pandas as pd
import numpy as np
suppliers = pd.read_csv("./northwind-mongo-master/suppliers _del_1.csv", usecols=["SupplierID", "CompanyName", "City"])
products = pd.read_csv("./northwind-mongo-master/products_del_1.csv", usecols=['ProductID', 'ProductName', 'SupplierID', 'CategoryID'])
categories = pd.read_csv("./northwind-mongo-master/categories1.csv", usecols=['CategoryID','CategoryName'], index_col="CategoryID")
```

```python
prod_cat = products.merge(categories, left_on="CategoryID", right_index=True)
```

```python
prod_cat_sup = prod_cat.merge(suppliers, left_on="SupplierID", right_on="SupplierID")
```
